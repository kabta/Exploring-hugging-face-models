```python

column_reduction_methods = [
    "Recursive feature elimination",
    "F-regression-percentile",
    "F-regression-Kbest",
    "f-regression-anova",
    "r-regression",
    "Mutual info regression"
]

row_reduction_methods = [
    "Random Under Sampling",
    "Cluster Centroids",
    "Near Miss"
]

model = [
    "Linear regression",
    "Ridge regression",
    "Lasso regression",
    "Elastic net regression",
    "Svr"
]

# Generate percentages from 90% to 10%
percentages = [f"{p}%" for p in range(90, 9, -10)]

# Create a DataFrame with all combinations
combinations = [(col_method, row_method, mdl, col_percentage, row_percentage)
                 for mdl in model
                 for col_method in column_reduction_methods
                 for row_method in row_reduction_methods
                 for col_percentage in percentages
                 for row_percentage in percentages]

# Create a DataFrame using the combinations
exp_list = pd.DataFrame(combinations, columns=['model','column_reduction_method', 'row_reduction_method',
                                         'FeatureSelection_percentage', 'RowSelection_percentage'])

# Add a numerical column 'Result #' starting from 1
exp_list['Result#'] = range(1, len(exp_list) + 1)

# Set 'Result #' column as the index and rename it to 'exp_Id'
exp_list.index.name = 'exp_Id'
```

# Feature reduction

```python

def feature_reduction(column_reduction_method, dataset, FeatureSelection_percentage):

    # Filter rows based on the specified conditions
    filtered_df = dataset[
        (dataset['column_reduction_method'] == column_reduction_method) &
        (dataset['FeatureSelection_percentage'] == FeatureSelection_percentage)
    ]

    return filtered_df


```

```python
# Iterate through the rows of the dataset and apply the feature_reduction function
for index, row in exp_list.iterrows():
    column_reduction_method = row['column_reduction_method']
    FeatureSelection_percentage = row['FeatureSelection_percentage']

    # Apply the feature_reduction function
    selected_data = feature_reduction(column_reduction_method, exp_list, FeatureSelection_percentage)

    # Display the selected dataset for the current iteration
    print(f"Selected data for exp_Id {index}:\n{selected_data}\n")
```
